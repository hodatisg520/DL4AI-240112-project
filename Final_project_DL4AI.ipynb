### Kết nối với GitHub và sử dụng Commit History trong Google Colab

Google Colab tích hợp sẵn với GitHub, cho phép bạn lưu trữ notebook của mình và quản lý các thay đổi thông qua hệ thống kiểm soát phiên bản Git.

#### 1. Lưu Notebook vào GitHub (Nếu chưa có trong Repo):

Nếu bạn đang làm việc trên một notebook mới hoặc một notebook chưa được liên kết với GitHub:

1.  **File > Save a copy in GitHub**.
2.  Bạn sẽ được yêu cầu cấp quyền truy cập Colab vào tài khoản GitHub của mình (nếu đây là lần đầu tiên).
3.  Chọn kho lưu trữ (repository) và chi nhánh (branch) mà bạn muốn lưu.
4.  Nhập tin nhắn commit và nhấn **OK**.

#### 2. Mở Notebook từ GitHub:

Nếu bạn muốn mở một notebook đã có trên GitHub:

1.  **File > Open notebook**.
2.  Chọn tab **GitHub**.
3.  Bạn có thể tìm kiếm kho lưu trữ công khai hoặc kho lưu trữ của mình.
4.  Chọn notebook bạn muốn mở.

#### 3. Thực hiện thay đổi và Commit:

Khi bạn đã mở một notebook từ GitHub:

1.  Thực hiện các thay đổi cần thiết trong notebook của bạn.
2.  Để lưu các thay đổi này trở lại GitHub, đi tới **File > Save a copy in GitHub** (hoặc **File > Save and commit** nếu bạn đã lưu một lần trước đó).
3.  Hộp thoại **Commit changes** sẽ xuất hiện.
4.  Nhập **Commit message** mô tả những thay đổi bạn đã thực hiện.
5.  (Tùy chọn) Chọn chi nhánh nếu bạn muốn commit vào một chi nhánh khác.
6.  Nhấn **OK**.

#### 4. Xem lịch sử Commit (Commit History):

Colab không có giao diện tích hợp trực tiếp để hiển thị lịch sử commit trong notebook như một IDE đầy đủ. Tuy nhiên, bạn có thể xem lịch sử commit của file trên GitHub:

1.  **File > View revisions** (hoặc **File > See revision history**).
2.  Thao tác này sẽ mở trình duyệt của bạn và điều hướng đến trang lịch sử commit của file đó trên GitHub.
3.  Tại đó, bạn có thể xem tất cả các commit, tác giả, tin nhắn commit và các thay đổi đã được thực hiện trong mỗi commit.
